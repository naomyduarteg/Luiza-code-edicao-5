Exercício 1

```python
import math

#1. 
try:
    n = float(input("Digite um número n positivo: "))
    print(f"A raíz quadrada de n é: {math.sqrt(n)}")
    if n<0:
        raise ValueError
except ValueError as VE:
    print(VE)
    print("O número escolhido não é positivo.")
```

Exercício 2

```python
#2. 
try:
    n = float(input("Digite um nominador: "))
    m = float(input("Digite um denominador: "))
    print(f"A divisão n por mé: {n/m}")
    if m==0:
        raise ZeroDivisionError
except ZeroDivisionError as ZDE:
    print(ZDE)
    print("O denominador é igual a zero. Você criou um buraco negro!")
```
